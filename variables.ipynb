```python
import os

variables = [
    "OPENAI_API_KEY",
    "OPENAI_BASE_URL",
    "ANTHROPIC_API_KEY",
    "ANTHROPIC_BASE_URL",
    "ANTHROPIC_CUSTOM_URL",
    "ANTHROPIC_CUSTOM_KEY",
    "OPENAI_CUSTOM_URL",
    "OPENAI_CUSTOM_KEY",
    "GEMINI_CUSTOM_URL",
    "GEMINI_CUSTOM_KEY"
]

for var in variables:
    value = os.environ.get(var)
    if value:
        print(f"{var}: {value}")
    else:
        print(f"{var}: Not Found")
```
